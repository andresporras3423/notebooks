<h3>Marco teórico</h3>
<ul>
<li>outliers: también conocidos como "valores atípicos" en estadística y análisis de datos son observaciones que se desvían tanto del resto de los datos que levantan sospechas sobre su validez o causan preocupación sobre si proceden de una distribución diferente.</li>
<li>estandarización: se trata un proceso estadístico utilizado para hacer que los datos sean más comparables y manejables, especialmente cuando involucran diferentes unidades o escalas.</li>
<li>categorización:  es el proceso de agrupar o clasificar datos en categorías o clases distintas. Este proceso es fundamental para organizar datos, simplificar el análisis y facilitar la comprensión de los patrones y tendencias en los datos. Por ejemplo, la variable edad podría categorizarse por rangos, por ejemplo; infancia, adolescencia, adultez, etc.</li>
<li>codificación: Este concepto se refiere al proceso de transformar datos de un formato a otro, generalmente con el objetivo de preparar los datos para un análisis más efectivo o para su uso en aplicaciones específicas como el aprendizaje automático. Por ejemplo, si tienes una variable categórica como "color" con etiquetas "rojo", "verde" y "azul", la codificación podría convertir esto en 1 para "rojo", 2 para "verde", y 3 para "azul".</li>
<li>folium: es una biblioteca en Python que se utiliza para crear mapas interactivos. Utiliza Leaflet.js, una biblioteca de JavaScript para mapas interactivos, y permite a los usuarios de Python integrar sus capacidades de mapeo directamente en un entorno de Python.</li>
<li>diagrama de cajas: también conocido como diagrama de caja y bigotes o boxplot, es una representación gráfica utilizada en estadística para mostrar la distribución de un conjunto de datos. Este tipo de gráfico es muy útil para visualizar la mediana, los cuartiles y los valores atípicos (outliers) de los datos. El diagram tiene los siguientes componentes:
    <ol>
        <li>caja central: representa el rango intercuartílico (RIC), que abarca el 50% central de los datos</li>
        <li>línea dentro de la caja: muestra la media del conjunto de datos.</li>
        <li>bigotes: estos muestran datos en un rango de 1.5 veces el RIC tanto en la parte inferior como superior del intercuartil. Por ejemplo, si el rango intercuartil va desde 25 hasta 45, entonces el RIC sería 45-25=20, y entonces los bigotes cubrirían el rango de -5 a 25, y de 45 a 75. </li>
        <li>puntos: representan los valores atípicos del conjunto de datos.</li>
    </ol>

</li>
<li>diagrama de violin: una herramienta de visualización de datos que combina elementos de un diagrama de cajas (boxplot) y una estimación de la densidad de probabilidad. Este tipo de gráfico es útil para visualizar la distribución de los datos de forma continua.</li>
</ul>

```python
import pandas as pd
csv_file_path = r"hotel_bookings.csv" # csv file path
df_hotels_booking = pd.read_csv(csv_file_path)  # read table in csv file
month_to_number = { # mapping month category to number
    "January": 1, "February": 2, "March": 3, "April": 4,
    "May": 5, "June": 6, "July": 7, "August": 8,
    "September": 9, "October": 10, "November": 11, "December": 12
}
df_hotels_booking["children"].fillna(0, inplace=True) # to replace nulls with 0 in the column "children"
df_hotels_booking["arrival_month_number"] = df_hotels_booking["arrival_date_month"].map(month_to_number) # encode month variable
df_hotels_booking["total_people"] = df_hotels_booking["adults"]+df_hotels_booking["children"]+df_hotels_booking["babies"] # create variable total people = adults + children + babies
df_hotels_booking['total_stay_nights'] = df_hotels_booking['stays_in_weekend_nights'] + df_hotels_booking['stays_in_week_nights'] # create variable total stay nights = stay week nights + stay weekend nights

continents_file_path = r"countries_data.csv" # csv file path

df_continents = pd.read_csv(continents_file_path) # tis file will alow to map continent name with country code.

# below line differentiate between south and north america to get more specific data for south america region.
df_continents["region"] = df_continents.apply(lambda row: row["region"] if row["region"] != "Americas" else ("South America" if row["intermediate-region"]=="South America" else "Nothern America"), axis=1)

df_continents = df_continents[["region","alpha-3"]]
df_continents.rename(columns={'region': 'continent', 'alpha-3': 'country'}, inplace=True)
df_hotels_booking = pd.merge(df_hotels_booking, df_continents, on='country', how='left')
```

```python
adr_min=df_hotels_booking["adr"].min()
adr_max=df_hotels_booking["adr"].max()
adr_range = adr_max-adr_min
adr_bin_size =  adr_range/5.0
adr_sum=df_hotels_booking["adr"].sum()
adr_bin_sum=adr_sum/5.0
#print(adr_sum)
#print(adr_bin_sum)

# Seleccionar y ordenar la columna
adr_sorted = df_hotels_booking["adr"].sort_values(ascending=False)

# Convertir a lista
adr_list = adr_sorted.tolist()

adr_bins=[]

adr_acc=0

adr_bin_limit=adr_bin_sum

for current_adr in adr_list:
    adr_acc += current_adr
    if(adr_acc>adr_bin_limit):
        adr_bins.append(current_adr)
        adr_bin_limit+=adr_bin_sum
        
adr_bins = [adr_max+1]+adr_bins
adr_bins[-1] = adr_min-1
adr_bins.sort()

labels = ['very low rate', 'low rate', 'average rate', 'high rate', 'very high rate']  # define category names
df_hotels_booking['adr_by_category'] = pd.cut(df_hotels_booking['adr'], bins=adr_bins, labels=labels, right=False)
df_hotels_booking['adr_by_category'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler = StandardScaler()
standard_variables =['adr std', 'children std','adults std','babies std', 'total_people std'] # name standarized variables
original_variables =['adr', 'children','adults','babies', 'total_people'] # name variables to be standarized
df_hotels_booking[standard_variables] = scaler.fit_transform(df_hotels_booking[original_variables]) 
df_hotels_booking[standard_variables].describe().T
```

```python
def filter_quantile(column): # remove extreme values to make violin diagram more understandable
    lower_bound = column.quantile(0.01)
    upper_bound = column.quantile(0.99)
    return column[(column >= lower_bound) & (column <= upper_bound)]

filtered_df_std =df_hotels_booking[standard_variables].apply(filter_quantile)

# Prepare the DataFrame for plotting
df_melted = filtered_df_std.melt()

# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='variable', y='value', data=df_melted)
plt.title("violin plot, variables estandarizadas")
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()
plt.show()
```

```python
def add_scatters(col):  # add scatter of the given column
    values, counts = np.unique(df_hotels_booking[col], return_counts=True)
    plt.scatter(values, counts, alpha=0.5, label=col, edgecolors='w')

#  create scatter plot
plt.figure(figsize=(10, 8))

# add scatter for each standarized column
for std in standard_variables:
    add_scatters(std)

plt.title('Diagrama de puntos de la frecuencia')
plt.xlabel('Valores Únicos')
plt.ylabel('Frecuencia')

plt.legend()

plt.grid(True)

plt.show()
```

```python
df_stats_standarized = pd.DataFrame(columns=["standard deviation", "mean","kurtosis", "skewness","media frequency"]) # list of normal distribution stats
# function add the numeric stats for the column that it receives as parameter
def stats_standarized(col):
    global df_stats_standarized
    column = df_hotels_booking[col]
    valor_min = column.min() 
    valor_max = column.max()  
    new_row = pd.DataFrame( {
        "standard deviation":round(column.std(),2),
        "mean":round(column.mean(),2),
        "kurtosis":round(column.kurtosis(),2),
        "skewness":round(column.skew(),2)
        ,"media frequency": column.value_counts().iloc[0]
    }, index=[col])
    df_stats_standarized = pd.concat([df_stats_standarized, new_row], axis=0)
# add new row for each standarized column
for std in standard_variables:
    stats_standarized(std)
    
print(df_stats_standarized)
```

<h3>Puntuación mejorada: Si su base de datos contiene datos georeferenciados, presente una visualización de mapa con la librería Folium.</h3>

```python
import folium
import json

country_counts_df = df_hotels_booking["country"].value_counts()
country_counts_df = country_counts_df.reset_index()
country_counts_df.columns = ['country', 'count'] # create dataframe with country code and their count of records
 
m = folium.Map(location=[20, 0], zoom_start=2)

with open('custom.geo.json', 'r', encoding='utf-8') as file:
    geojson_data = json.load(file) # load geojson file to get borders between countries

def custom_color_scale(count): # set color by count range
        if count == 0:
            return '#FFFFFF' # white color for countries with no records
        elif count < 2000:
            return '#C2E699'  
        elif count < 4000:
            return '#78C679' 
        elif count < 8000:
            return '#31A354'  
        elif count < 16000:
            return '#006837'  
        else:
            return '#003300' 
folium.GeoJson(
    geojson_data,
    name='choropleth',
    style_function=lambda feature: {
        'fillColor': custom_color_scale(
            country_counts_df.loc[country_counts_df['country'] == feature['properties']['adm0_a3'], 'count'].values[0]
            if feature['properties']['adm0_a3'] in country_counts_df['country'].values
            else 0  # if match is found then use count, otherwise set count of records as 0
        ),
        'fillOpacity': 0.7,
        'color': 'black',
        'weight': 1,
    },
    tooltip='Country: {0}, Count: {1}'.format(country_counts_df['country'], country_counts_df['count']),
).add_to(m)

m
```

<h3>Análisis del mapa</h3>
<p>La visualización refleja a Portugal como el país con más datos, también podemos ver que los países con más registros se encuentran en Europa. También podemos observar que algunos países no tienen ningún registro, entre ellos algunos países grandes como Canadá.</p>
