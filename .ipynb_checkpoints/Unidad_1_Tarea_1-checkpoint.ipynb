##  2) Realizar las tablas de distribuciones de frecuencias y gráficas de cada una de las variables categóricas, colocando para cada una, su correspondiente interpretación descriptiva.

```python
def clasificar_variable(tipo):
    if pd.api.types.is_numeric_dtype(tipo):
        return 'Numérica'
    elif pd.api.types.is_string_dtype(tipo):
        return 'Categórica'
    else:
        return 'Otra'
categories = [cat for cat in df.columns if clasificar_variable(df[cat].dtype)=='Categórica']
def generate_graph(variable_categorica):
  # Contar frecuencias de cada categoría
  frecuencias = df[variable_categorica].value_counts()
  # Crear el gráfico de barras
  plt.figure(figsize=(6, 4))
  plt.bar(frecuencias.index, frecuencias.values, color='skyblue')
  plt.xlabel('Categoría')
  plt.ylabel('Frecuencia')
  plt.title(f'Frecuencia de {variable_categorica}')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()  # Ajustar el diseño para evitar el corte de etiquetas
  plt.show()

for categoria in categories:
  generate_graph(categoria)
```

## 6) Analizar las asociaciones estado de fidelidad del cliente, reservación y tipo de vehículo, siguiendo las indicaciones:

### Realiza una tabla de contingencia para cada una de las asociaciones (deben salir tres tablas), coloque observaciones relevantes frente a las posibles relaciones que encuentre entre categorías.

```python
class contingency_data():
  # Método constructor (se llama cuando se crea una instancia de la clase)
  def __init__(self, category1, category2, table):
    self.category1 = category1  # Atributo nombre
    self.category2 = category2      # Atributo edad
    self.table = table
    self.chi2_1, self.p_1, self.dof_1, self.ex_1 = chi2_contingency(table)

  def chi_square_meaning(self):
    threshold_significance = 0.05
    threshold_weak_evidence = 0.1

    if self.p_1 < threshold_significance:
        print(f"Con un valor de p igual a {self.p_1} para la Tabla {con.category1} vs {con.category2}, hay una asociación significativa.")
    elif self.p_1 <= threshold_weak_evidence:
        print(f"Con un valor de p igual a {self.p_1} para la Tabla {con.category1} vs {con.category2}, hay evidencia de una asociación débil, pero no significativa.")
    else:
        print(f"Con un valor de p igual a {self.p_1} para la Tabla {con.category1} vs {con.category2}, no hay evidencia suficiente para decir que hay una asociación.")


print(categories)
list_contingency = []
def generate_contingency_table(cat1, cat2):
  contingency_table = pd.crosstab(df[cat1], df[cat2])
  list_contingency.append(contingency_data(cat1,cat2, contingency_table))


for i in range(0,len(categories)):
  for j in range(i+1,len(categories)):
    generate_contingency_table(categories[i],categories[j])

for con in list_contingency:
  print(f"\n\nTABLA DE CONTINGENCIA - {con.category1} vs {con.category2}:")
  display(con.table)
```

## 7) Analizar la relación entre número de viajeros, calificación y expectativa de duración del viaje, siguiendo las indicaciones:

### Diseñe gráficos de boxplot o violín plot para analizar si se detectan diferencias en las distribuciones de viajeros, calificación y expectativa de duración del viaje según la fidelidad del cliente. (Debe obtener tres gráficas)

```python
def violin_comparison(loyalty_status):
  # Escalar las columnas seleccionadas
  comparison_labels = ["Number_of_Riders", "Average_Ratings", "Expected_Ride_Duration"]
  df_violin = df[comparison_labels][df["Customer_Loyalty_Status"]==loyalty_status].copy()
  df_violin
  scaler = StandardScaler()
  df_violin = scaler.fit_transform(df_violin)
  # Crear el gráfico de violín horizontal
  plt.figure(figsize=(8, 5))
  sns.violinplot(data=df_violin, orient='h')
  # Ajustar título, etiquetas y ticks del eje y
  plt.title(f"comparison variables distribution for {loyalty_status} loyalty status")
  plt.xlabel('Value')
  plt.ylabel('Variables')
  plt.yticks(ticks=range(3), labels=comparison_labels)  # Configurar las etiquetas basadas en los nombres de las columnas
  plt.grid(True)
  plt.show()

unique_values = df['Customer_Loyalty_Status'].unique()

# Iterar sobre cada valor único
for value in unique_values:
  violin_comparison(value)
```

```python
# ANOVA de una sola vía para comparar las medias en diferentes niveles de fidelidad del cliente

def anova_test(variable):
    # Realizar ANOVA de una sola vía
    model = ols(f'{variable} ~ Customer_Loyalty_Status', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Mostrar resultados de la prueba ANOVA
    print(f"ANOVA Results for {variable}:\n")
    print(anova_table)

    # Comprobar significancia
    if anova_table['PR(>F)'].iloc[0]< 0.05:
        print(f"\nLa fidelidad del cliente tiene una influencia significativa en {variable}.\n")
        # Si hay significancia, realizar prueba Post Hoc (Tukey HSD)
        tukey = pairwise_tukeyhsd(df[variable], df['Customer_Loyalty_Status'])
        print(tukey)
    else:
        print(f"\nNo se encontraron diferencias significativas para {variable}.\n")

# Realizar pruebas ANOVA para las tres variables
anova_test("Number_of_Riders")
anova_test("Average_Ratings")
anova_test("Expected_Ride_Duration")
```

<h2>Interpretación análisis ANOVA</h2>
<p>En los tres análisis de varianza (ANOVA), no se hallaron diferencias estadísticamente significativas para ninguna de las variables en relación al estado de lealtad del cliente (Customer_Loyalty_Status). Esto sugiere que el estatus de lealtad no tiene un impacto significativo en el número de pasajeros, las calificaciones promedio ni la duración esperada de los viajes.</p>
