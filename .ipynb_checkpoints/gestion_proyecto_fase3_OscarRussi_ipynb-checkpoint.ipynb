# Ejercicio 3.

```python
def group_by_metricas(df_, cols):
  df_rendimiento_ = df_.groupby(cols) \
    .agg(
        Total_Impresiones=("Impresiones", "sum"),
        Total_Clics=("Clics", "sum"),
        Total_Conversiones=("Conversiones", "sum")
    ).reset_index()
  # Evitar divisiones por cero
  df_rendimiento_ = df_rendimiento_[df_rendimiento_["Total_Impresiones"] > 0]
  df_rendimiento_ = df_rendimiento_[df_rendimiento_["Total_Clics"] > 0]
  # Calcular métricas
  df_rendimiento_["CTR (%)"] = (df_rendimiento_["Total_Clics"] / df_rendimiento_["Total_Impresiones"]) * 100
  df_rendimiento_["Tasa de Conversión (%)"] = (df_rendimiento_["Total_Conversiones"] / df_rendimiento_["Total_Clics"]) * 100
  df_rendimiento_["Conversión por Impresión (%)"] = (df_rendimiento_["Total_Conversiones"] / df_rendimiento_["Total_Impresiones"]) * 100
  df_rendimiento_[["CTR (%)", "Tasa de Conversión (%)", "Conversión por Impresión (%)"]] = \
  df_rendimiento_[["CTR (%)", "Tasa de Conversión (%)", "Conversión por Impresión (%)"]].round(2)
  return df_rendimiento_.sort_values(by="Tasa de Conversión (%)", ascending=False)

df_rendimiento = group_by_metricas(df_merge_campanhas_metricas, ["ID_Campaña", "Nombre"])
df_rendimiento
```

Además, se posiciona como la tercera mejor en cuanto a tasa de conversión, lo que indica un equilibrio sólido entre alcance, interacción y efectividad.

```python
# "Edad", "Género", "Intereses", "Ubicación ", "Comportamiento"
def pie_chart_agrupado(df, columnas, top_n=8):
    conteo = df[columnas[0] if len(columnas)==1 else columnas].value_counts()
    top = conteo.head(top_n)
    otros = conteo.iloc[top_n:].sum()

    # Definir colores para cada categoría
    colors = sns.color_palette("Set2", n_colors=top_n)

    labels = list(top.index) + ["Otros"]
    sizes = list(top.values) + [otros]

    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, colors=colors + ['lightgrey'], autopct="%1.1f%%", startangle=90, shadow=True)
    plt.title(f"Segmentación por {', '.join(columnas)}", pad=30)
    plt.axis("equal")
    plt.show()

pie_chart_agrupado(df_audiencias, ["Intereses"])
```

# Ejercicio 5. Monitoreo y control.

```python
from IPython.display import HTML, display

def limpiar_porcentaje(df_,col):
  col_limpia = (
  df_[col]
    .astype(str)
        .str.replace('%', '', regex=False)
        .str.replace(',', '.', regex=False)
  )
  return pd.to_numeric(col_limpia, errors='coerce')


# 2. Eliminar el símbolo de dólar y convertir a float (reemplazar coma de miles si aplica)
def limpiar_dinero(df_,col):
  col_limpia = (
        df_[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)  # Aquí eliminamos coma de miles, no usamos punto
  )
  return pd.to_numeric(col_limpia, errors='coerce')

# 1. Eliminar el símbolo de porcentaje y convertir a float (reemplazar coma por punto)

def limpiar_metricas(df_):
  # 4. Agrupar por campaña y calcular promedio SOLO para columnas específicas
  # 3. Aplicamos la limpieza a las columnas necesarias
  df_['Tasa_Clics'] = limpiar_porcentaje(df_,'Tasa_Clics')
  df_['Tasa_Conversión'] = limpiar_porcentaje(df_,'Tasa_Conversión')
  df_['Tasa_rebote'] = limpiar_porcentaje(df_,'Tasa_rebote')
  df_['Ingreso_Generado'] = limpiar_dinero(df_,'Ingreso_Generado')
  return df_

def promedio_metricas(df_, group_cols):
  columnas_metricas = [
    "Clics", "Impresiones", "Tasa_Clics", "Conversiones",
    "Tasa_Conversión", "Ingreso_Generado", "Rebotes", "Tasa_rebote"
    ]
  df_ = limpiar_metricas(df_)
  return df_.groupby(group_cols)[columnas_metricas].mean()

promedios_por_campaña = promedio_metricas(df_metricas,["ID_Campaña"])

# Calcular la media global de todas las campañas (por cada métrica)
media_global = promedios_por_campaña.mean().to_frame().T  # Transponemos para que quede como una fila

# Renombrar el índice (la fila) como 'promedio'
media_global.index = ['promedio']
display(HTML("<h2>Promedio métricas por campaña</h2>"))
# Mostrar resultado
display(media_global)
```

```python
def sum_metricas(df_, groupby_col=None):
    df_['Ingreso_Generado'] = limpiar_dinero(df_, 'Ingreso_Generado')
    metricas_rendimiento = ["Clics", "Impresiones", "Conversiones", "Ingreso_Generado", "Rebotes"]
    if groupby_col:
      # Agrupar por columna si se especifica
      suma = df_.groupby(groupby_col)[metricas_rendimiento].sum()
    else:
      # Sumar directamente si no se quiere agrupar
      suma = df_[metricas_rendimiento].sum().to_frame(name="Suma Total")
    return suma

display(HTML("<h2>Suma total de las principales métricas</h2>"))
# Mostrar resultado
display(sum_metricas(df_metricas))
```

## Canales de marketing que generan mayor retorno de inversión (ROI).

```python
# 2. Eliminar el símbolo de dólar y convertir a float (reemplazar coma de miles si aplica)
df_merge_campanhas_metricas["Presupuesto"] = limpiar_dinero(df_merge_campanhas_metricas,"Presupuesto")
df_merge_campanhas_metricas["Ingreso_Generado"] = limpiar_dinero(df_merge_campanhas_metricas,"Ingreso_Generado")

def roi_agrupado(df_, cols):
  roi_agrupado = (
  df_
  .groupby(cols)[["Ingreso_Generado", "Presupuesto"]]
  .sum()
  .assign(ROI=lambda df: (df["Ingreso_Generado"]-df["Presupuesto"]) / df["Presupuesto"])
  .reset_index()
  )
  return roi_agrupado
roi_por_plataforma=roi_agrupado(df_merge_campanhas_metricas, "Plataforma")
roi_por_plataforma.sort_values(by="ROI", ascending=False)
```

## Campañas que deberían mantenerse, que necesitan ajustes adicionales, que deberían descartarse.

```python
roi_por_campaña = roi_agrupado(df_merge_campanhas_metricas, "ID_Campaña")
display(HTML("<h2>Campañas que deberían mantenerse</h2>"))
roi_por_campaña[roi_por_campaña["ROI"]>0.5].sort_values(by="ROI", ascending=False)
```

```python
display(HTML("<h2>Campañas que necesitan ajustes</h2>"))
roi_por_campaña[(roi_por_campaña["ROI"] <= 0.5) & (roi_por_campaña["ROI"] >= -0.5)].sort_values(by="ROI", ascending=False)
```

```python
display(HTML("<h2>Campañas que deberían descartarse</h2>"))
roi_por_campaña[roi_por_campaña["ROI"] <-0.5].sort_values(by="ROI", ascending=False)
```

A lo largo de esta fase del proyecto se llevó a cabo un análisis integral de campañas publicitarias y segmentos de audiencia, utilizando diversas métricas clave para evaluar su rendimiento. A partir de los resultados obtenidos, se pueden establecer las siguientes conclusiones:

La relación uno a muchos entre campañas, métricas y audiencias genera duplicados al unir las tres tablas, lo que sugiere que para ciertos análisis es más apropiado realizar fusiones por separado entre campañas y cada uno de los otros dos conjuntos.

El análisis de métricas clave reveló que algunas campañas, como la identificada con el ID 27, destacaron por su equilibrio entre alcance e impacto, al liderar en conversiones totales y posicionarse entre las mejores en tasa de conversión y valor por impresión.

El estudio de los segmentos de audiencia permitió identificar un total de 114 combinaciones únicas. Sin embargo, la mayoría presentó un número reducido de registros, lo cual limita la confiabilidad de las inferencias en algunos casos.

Se identificaron patrones relevantes en el comportamiento de las audiencias: por ejemplo, los segmentos de hombres entre 25 y 34 años en determinadas ubicaciones mostraron altos valores de CTR, mientras que el interés por tecnología y negocios fue común entre los segmentos con mayores ingresos generados.

Finalmente, el análisis de ROI permitió categorizar las campañas según su efectividad financiera. Mientras que algunas lograron retornos superiores al 200%, otras presentaron rendimientos cercanos a cero o negativos, recomendando así su eliminación o replanteamiento estratégico.
