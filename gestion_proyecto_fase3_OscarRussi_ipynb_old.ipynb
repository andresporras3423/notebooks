## Métricas de una campaña específica.


```python
print("Número de métricas por campaña:")
df_merge_campanhas_metricas.groupby(["ID_Campaña", "Nombre"]).size().reset_index(name="Conteo").sort_values(by="Conteo", ascending=False)
```

```python
def tabla_conteo_dia_madre(df, col_group):
  print(f"Número de registros por {col_group} para la campaña: Campaña Día de la Madre 2025")
  return df_merged_final[df_merged_final["ID_Campaña"]==30].groupby(["ID_Campaña",col_group])\
  .size().reset_index(name="Conteo").sort_values(by="Conteo", ascending=False)


tabla_conteo_dia_madre(df_merge_campanhas_metricas, "ID_Métrica")
```

## Métricas de clics e impresiones para todas las campañas.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def create_bar(df, colx, coly):
    # Agrupación
    df_resumen = df.groupby(colx).agg(Suma=(coly, "sum")).reset_index()
    df_resumen = df_resumen[df_resumen["Suma"] > 0]
    # Plot
    plt.figure(figsize=(10, 8))
    sns.barplot(data=df_resumen, x=colx, y="Suma", errorbar=None)

    plt.title(f"Suma de {coly} por {colx}")
    plt.xlabel(colx)
    plt.ylabel(f"Suma de {coly}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


# Llamada
create_bar(df_merge_campanhas_metricas, "ID_Campaña", "Clics")
```

```python
create_bar(df_merge_campanhas_metricas, "ID_Campaña", "Impresiones")
```

## El rendimiento promedio de conversión por tipo de campaña.

```python
df_conversion = df_merge_campanhas_metricas.groupby("Tipo_Campaña") \
    .agg(Total_Clics=("Clics", "sum"), Total_Conversiones=("Conversiones", "sum")) \
    .reset_index()

# Calcular tasa de conversión como porcentaje
df_conversion["Tasa_Conversión (%)"] = (df_conversion["Total_Conversiones"] / df_conversion["Total_Clics"]) * 100
df_conversion.sort_values(by="Tasa_Conversión (%)", ascending=False)
```

# Ejercicio 3.

```python
df_rendimiento = df_merge_campanhas_metricas.groupby(["ID_Campaña", "Nombre"]) \
    .agg(
        Total_Impresiones=("Impresiones", "sum"),
        Total_Clics=("Clics", "sum"),
        Total_Conversiones=("Conversiones", "sum")
    ).reset_index()

# Evitar divisiones por cero
df_rendimiento = df_rendimiento[df_rendimiento["Total_Impresiones"] > 0]
df_rendimiento = df_rendimiento[df_rendimiento["Total_Clics"] > 0]

# Calcular métricas
df_rendimiento["CTR (%)"] = (df_rendimiento["Total_Clics"] / df_rendimiento["Total_Impresiones"]) * 100
df_rendimiento["Tasa de Conversión (%)"] = (df_rendimiento["Total_Conversiones"] / df_rendimiento["Total_Clics"]) * 100
df_rendimiento["Conversión por Impresión (%)"] = (df_rendimiento["Total_Conversiones"] / df_rendimiento["Total_Impresiones"]) * 100

# Redondear valores
df_rendimiento[["CTR (%)", "Tasa de Conversión (%)", "Conversión por Impresión (%)"]] = \
    df_rendimiento[["CTR (%)", "Tasa de Conversión (%)", "Conversión por Impresión (%)"]].round(2)
df_rendimiento.sort_values(by="Tasa de Conversión (%)", ascending=False)
```

Basado en las métricas evaluadas durante el ejercicio 2, la campaña con ID 27 muestra el mejor desempeño general. No solo registra el mayor número total de conversiones, sino que también comparte el mejor valor de conversión por impresión junto a las campañas con ID 3 y 2. Además, se posiciona como la tercera mejor en cuanto a tasa de conversión, lo que indica un equilibrio sólido entre alcance, interacción y efectividad.
